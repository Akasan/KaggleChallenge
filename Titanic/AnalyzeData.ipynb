# Preprocessing

```python
def to_number(df):
    df.Sex = df.Sex.map({"male": 0, "female": 1})  # [Sex] male: 0, female: 1
    df.Embarked = df.Embarked.map({"S": 0, "C": 1, "Q": 2})
    return df
```

```python
train_df = to_number(train_df)
```
